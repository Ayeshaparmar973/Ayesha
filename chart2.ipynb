```python
import pandas as pd
data = pd.read_csv("/content/traffic.csv (2) (2).zip")
print(data.head())
print(data.columns)
import matplotlib.pyplot as plt

# Example: Plot traffic volume by day
plt.bar(data['DateTime'], data['Vehicles'])
plt.xlabel('DateTime')
plt.ylabel(' Vehicles')
plt.title(' Vehicles by DateTime')
plt.show()
```
